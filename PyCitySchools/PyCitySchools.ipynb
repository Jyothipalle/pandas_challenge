```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"


# Read School and Student Data File and store into Pandas DataFrames
school_data_df = pd.read_csv(school_data_to_load)
student_data_df= pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])

```

```python
#calcualte the total number of school
schools_total = school_data_complete_df["school_name"].nunique()
schools_total
```

```python
#calculate the total students
total_students = school_data_complete_df["Student ID"].nunique()
total_students
```

```python
#calculate the totalbudget
total_budget = school_data_df["budget"].sum()
total_budget
```

```python
#calculate the average maths score
avg_maths_score = school_data_complete_df["math_score"].mean()
avg_maths_score
```

```python
#calculate the average reading score
avg_reading_score = school_data_complete_df["reading_score"].mean()
avg_reading_score
```

```python
# calculate the percentage passing maths

passing_maths= school_data_complete_df.loc[school_data_complete_df["math_score"] >= 70]
number_passing_maths=passing_maths["Student ID"].count()
percentage_passing_maths= (number_passing_maths/total_students)*100
percentage_passing_maths
```

```python
# calculate the percentage passing reading
passing_reading = school_data_complete_df.loc[school_data_complete_df["reading_score"] >= 70]
number_passing_reading=passing_reading["Student ID"].count()
percentage_passing_reading = (number_passing_reading/total_students)*100
percentage_passing_reading
```

```python
#overall percentage passing
passing_overall = school_data_complete_df[(school_data_complete_df["reading_score"] >= 70) & (school_data_complete_df["math_score"] >= 70)]["Student ID"].count()

percentage_passing_overall = (passing_overall/total_students)*100
percentage_passing_overall
```

```python
#creating a DataFrame to hold the above results
school_df = pd.DataFrame({'Total Schools': schools_total,
                          'Total Students': total_students,
                          'Total Budget' :f"${total_budget:,.2f}",
                          'Average Math Score':avg_maths_score,
                          'Average Reading Score':avg_reading_score,
                          '%Passing Maths':percentage_passing_maths,
                          '%Passing Reading':percentage_passing_reading,
                          '% Overall Passing':percentage_passing_overall
                          },index=[0])
school_df
```

```python
#grouping the columns with school_name and summarising the details 
school_group_df = school_data_complete_df.groupby("school_name")
sch_type=school_group_df["type"].first()
sch_tot_student=school_group_df["student_name"].count()
sch_budget = school_group_df["budget"].mean()
sch_per_student_budget = school_group_df["budget"].mean() /school_group_df["student_name"].count()
sch_avg_maths_score = school_group_df["math_score"].mean() 
sch_avg_reading_score = school_group_df["reading_score"].mean()
sch_math_pass = school_data_complete_df[school_data_complete_df['math_score'] >= 70].groupby("school_name")['Student ID'].count()/sch_tot_student*100
sch_reading_pass = school_data_complete_df[school_data_complete_df['reading_score'] >= 70].groupby("school_name")['Student ID'].count()/sch_tot_student*100
sch_overall_pass = school_data_complete_df[(school_data_complete_df['reading_score'] >= 70) & (school_data_complete_df['math_score'] >= 70)].groupby("school_name")["student_name"].count()/sch_tot_student*100

#creating the dataframe to hold the values
sch_summary_df = pd.DataFrame({
                          'School Type' :sch_type,
                          'Total Students': sch_tot_student,
                          'Total School Budget' :sch_budget,
                          'Per Student Budget' :sch_per_student_budget,
                          'Average Math Score':sch_avg_maths_score,
                          'Average Reading Score':sch_avg_reading_score,
                          '% Passing Math':sch_math_pass,
                          '% Passing Reading':sch_reading_pass,
                          '% Overall Passing':sch_overall_pass
                          })

# Copying to new dataframe without formating to use later
sch_summary_copy_df = sch_summary_df.copy()
# Formatting 
sch_summary_df["Total School Budget"]=sch_summary_df["Total School Budget"].map("${:,.2f}".format)
sch_summary_df["Per Student Budget"]=sch_summary_df["Per Student Budget"].map("${:,.2f}".format)
# Changing index label to None
sch_summary_df.index.name= None
sch_summary_df
```

```python
# Sorting the dataframe to get Top performing schools(By % Overall Passing)
top_performing_sch=sch_summary_df.sort_values("% Overall Passing", ascending=False)
top_performing_sch
```

```python
#Sorting the dataframe by ascending to get Bottom Performing Schools (By % Overall Passing)

bottom_performing_sch=sch_summary_df.sort_values("% Overall Passing", ascending=True).head()
bottom_performing_sch
```

```python
# Getting average Math Scores by Grade and creating a series

math_9th=school_data_complete_df.loc[school_data_complete_df['grade'] == "9th"].groupby("school_name")['math_score'].mean()
math_10th=school_data_complete_df.loc[school_data_complete_df['grade'] == "10th"].groupby("school_name")['math_score'].mean()
math_11th=school_data_complete_df.loc[school_data_complete_df['grade'] == "11th"].groupby("school_name")['math_score'].mean()
math_12th=school_data_complete_df.loc[school_data_complete_df['grade'] == "12th"].groupby("school_name")['math_score'].mean()

#Creating a DataFrame to hold the values
maths_score_df=pd.DataFrame({
                            "9th" : math_9th,
                            "10th" : math_10th,
                            "11th" : math_11th,
                            "12th" : math_12th})
# Changing index label to None
maths_score_df.index.name=None

maths_score_df
```

```python
#Getting average Reading Scores by Grade and creating a series 

reading_9th=school_data_complete_df.loc[school_data_complete_df['grade'] == "9th"].groupby("school_name")['reading_score'].mean()
reading_10th=school_data_complete_df.loc[school_data_complete_df['grade'] == "10th"].groupby("school_name")['reading_score'].mean()
reading_11th=school_data_complete_df.loc[school_data_complete_df['grade'] == "11th"].groupby("school_name")['reading_score'].mean()
reading_12th=school_data_complete_df.loc[school_data_complete_df['grade'] == "12th"].groupby("school_name")['reading_score'].mean()

#creating a DataFrame to hold the values
reading_score_df=pd.DataFrame({
                            "9th" : reading_9th,
                            "10th" : reading_10th,
                            "11th" : reading_11th,
                            "12th" : reading_12th})
# Changing index label to None
reading_score_df.index.name=None

reading_score_df
```

```python
#creating bins to group school spending. Including the Average Math Score,Average Reading Score,% Passing Math,% Passing Reading,Overall Passing Rate

sch_bins = [0,584,629,644,675]

bin_group_names = ['<$585', '$585-630', '$630-645', '$645-675']
# Adding new column to school summary copy data frame
# Bin by per student budget using PD Cut

sch_summary_copy_df['Spending Ranges (Per Student)'] = pd.cut(sch_summary_copy_df['Per Student Budget'],
                                                         sch_bins,
                                                         labels=bin_group_names,
                                                         include_lowest=True)
# Group by 
school_spending_summary=sch_summary_copy_df.groupby('Spending Ranges (Per Student)')

school_spending_avg_math_score=school_spending_summary["Average Math Score"].mean()
school_spending_avg_reading_score=school_spending_summary["Average Reading Score"].mean()
school_spending_math_pass = school_spending_summary["% Passing Math"].mean()
school_spending_reading_pass = school_spending_summary["% Passing Reading"].mean()
school_spending_overall_pass = school_spending_summary["% Overall Passing"].mean()

# Create dataframe to hold values
school_spending_group_df=pd.DataFrame({
                            "Average Math Score" : school_spending_avg_math_score,
                            "Average Reading Score" : school_spending_avg_reading_score,
                            "% Passing Math" : school_spending_math_pass,
                            "% Passing Reading" : school_spending_reading_pass,
                            "% Overall Passing" : school_spending_overall_pass})

# Formatting the output to keep 2 decimals

school_spending_group_df["Average Math Score"]=school_spending_group_df["Average Math Score"].map("{:.2f}".format)
school_spending_group_df["Average Reading Score"]=school_spending_group_df["Average Reading Score"].map("{:.2f}".format)
school_spending_group_df["% Passing Math"]=school_spending_group_df["% Passing Math"].map("{:.2f}".format)
school_spending_group_df["% Passing Reading"]=school_spending_group_df["% Passing Reading"].map("{:.2f}".format)
school_spending_group_df["% Overall Pass"]=school_spending_group_df["% Overall Passing"].map("{:.2f}".format)


school_spending_group_df
```

```python
#creating bins based on school size. Including the Average Math Score,Average Reading Score,% Passing Math,% Passing Reading,Overall Passing Rate

sch_size_bins = [0,999, 1999, 5000]

bin_size_names = ['Small(<1000)', 'Medium(1000-2000)', 'Large(2000-5000)']

# Adding new column to school summary copy data frame
# Bin by school size using PD Cut

sch_summary_copy_df['School Size'] = pd.cut(sch_summary_copy_df['Total Students'],
                                                         sch_size_bins,
                                                         labels=bin_size_names,
                                                         include_lowest=True)
# Group by
school_size_summary=sch_summary_copy_df.groupby('School Size')


school_size_avg_math_score=school_size_summary["Average Math Score"].mean()
school_size_avg_reading_score=school_size_summary["Average Reading Score"].mean()
school_size_math_pass = school_size_summary["% Passing Math"].mean()
school_size_reading_pass = school_size_summary["% Passing Reading"].mean()
school_size_overall_pass = school_size_summary["% Overall Passing"].mean()

# Create dataframe to hold values
school_size_group_df=pd.DataFrame({
                            "Average Math Score" : school_size_avg_math_score,
                            "Average Reading Score" : school_size_avg_reading_score,
                            "% Passing Math" : school_size_math_pass,
                            "% Passing Reading" : school_size_reading_pass,
                            "% Overall Passing" : school_size_overall_pass})
school_size_group_df
```

```python
#creating bins based on school type. Including the Average Math Score,Average Reading Score,% Passing Math,% Passing Reading,Overall Passing Rate

school_type_summary=sch_summary_copy_df.groupby('School Type')

school_type_avg_math_score=school_type_summary["Average Math Score"].mean()
school_type_avg_reading_score=school_type_summary["Average Reading Score"].mean()
school_type_math_pass = school_type_summary["% Passing Math"].mean()
school_type_reading_pass = school_type_summary["% Passing Reading"].mean()
school_type_overall_pass = school_type_summary["% Overall Passing"].mean()

# Create dataframe to hold values

school_type_group_df=pd.DataFrame({
                            "Average Math Score" : school_type_avg_math_score,
                            "Average Reading Score" : school_type_avg_reading_score,
                            "% Passing Math" : school_type_math_pass,
                            "% Passing Reading" : school_type_reading_pass,
                            "% Overall Passing" : school_type_overall_pass})

school_type_group_df
```
